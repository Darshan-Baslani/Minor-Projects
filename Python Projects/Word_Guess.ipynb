A word guess game. You have to guess 5 words correctly in order to win. You have total 10 tries. If you guess a wrong character you would have to rewrite the whole word again. Maximum score one could obtain is 10. 1 point would be deducted with each wrong prediction.

```python
import sys
```

```python
def isApple(score,attempts_left):
    print('Guess the word')
    while True:
        print('A _ _ l _')
        ch = input('Enter prediction: ')
        if ch == 'p':
            print('A p _ l _')
            ch = input('Enter next prediction: ')
            if ch == 'p':
                print('A p p l _')
                ch = input('Enter next prediction: ')
                if ch == 'e':
                    print('Congrats you have guessed the first word Apple')
                    return score,attempts_left
                else:
                    score -= 1
                    attempts_left -= 1
                    if attempts_left == 0:
                        return score,attempts_left
                    else:
                        print(f'Your score is {score} with {attempts_left} attempts left')
                        print('You have to begin from the start of the word T_T')
            else:
                score -= 1
                attempts_left -= 1
                if attempts_left == 0:
                    return score,attempts_left
                else:
                    print(f'Your score is {score} with {attempts_left} attempts left')
                    print('You have to begin from the start of the word T_T')
        else:
            score -= 1
            attempts_left -= 1
            if attempts_left == 0:
                return score,attempts_left
            else:
                print(f'Your score is {score} with {attempts_left} attempts left')
                print('You have to begin from the start of the word T_T')                        
```

```python
score = 10
attempts_left = 10
words = ['apple', 'mango', 'boy', 'girl', 'google']
i=0

print('Guess all 5 words within 10 tries to win the game.')

#For Apple
print(f'Your score is {score} with {attempts_left} attempts left')
score, attempts_left = isApple(score,attempts_left) 
if attempts_left == 0:
    print('You have lost the game :(')
    sys.exit()
    
#For Mango
print(f'Your score is {score} with {attempts_left} attempts left')
score, attempts_left = isMango(score,attempts_left) 
if attempts_left == 0:
    print('You have lost the game :(')
    sys.exit()
    
#For Boy
print(f'Your score is {score} with {attempts_left} attempts left')
score, attempts_left = isBoy(score,attempts_left) 
if attempts_left == 0:
    print('You have lost the game :(')
    sys.exit()
    
#For Girl
print(f'Your score is {score} with {attempts_left} attempts left')
score, attempts_left = isGirl(score,attempts_left) 
if attempts_left == 0:
    print('You have lost the game :(')
    sys.exit()
    
#For Google
print(f'Your score is {score} with {attempts_left} attempts left')
score, attempts_left = isGoogle(score,attempts_left) 
if attempts_left == 0:
    print('You have lost the game :(')
else:
    print(f"Congrats you've won the game with the score of {score} :D")
```
